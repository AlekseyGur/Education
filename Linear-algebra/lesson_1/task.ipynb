### Задание 5

Установите, считается ли линейным подпространством:

а) совокупность всех векторов трёхмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, считающиеся линейными комбинациями данных векторов $\{u_{1}, u_{2} \ldots, u_{n}\}$.

### Решение 5

Чтобы пространство считать линейным, в нём должны работать операции сложения и умножения векторов.

а. 

$v_{1} = (0; y_{1}; z_{1})$

$v_{2} = (x_{2}; 0; z_{2})$

Сложение:

$v_{1} + v2_{2} = (0; y_{1}; z_{1}) + (x_{2}; 0; z_{2}) = (x_{2}; y_{1}; z1+z_{2})$

Если все значения x2, y1, z1, z2 отличны от нуля, то результат не лежит в том же множестве, что и исходные векторы, свойство линейности нарушается.

б.

$v_{1} = a_{1}*u_{1} + a_{2}*u_{2} + a_{3}*u_{3}$

$v_{2} = b_{1}*u_{1} + b_{2}*u_{2} + b_{3}*u_{3}$

Сложение:

$v_{1} + v_{2} = (a_{1}+ b_{1})*u_{1} + (a_{2}+b_{2})*u_{2} + (a_{3}+b_{3})*u_{3}$

Умножение:

$v_{1} * c = a_{1}*u_{1}*c + a_{2}*u_{2}*c + a_{3}*u_{3}*c$

Обе операции выполняются. Значит, это линейное подпространство

### Задание 2
Найдите нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

### Решение 2

Нормы:<br>
x = $||(4, 2, 4)|| = \sqrt{4^{2} + 2^{2} + 4^{2}} = 6$<br>
y = $||(12, 3, 4)|| = \sqrt{12^{2} + 3^{2} + 4^{2}} = 13$

Находим угол между векторами:

$$cos\varphi = \frac{(x,y)}{||x|| * ||y||} = \frac{(4*12 + 2*3 + 4*4)}{(6*13)} = 0,897$$

$$\varphi = 26,2 ^\circ$$

```python
import math
x = np.array([4, 2, 4])
y = np.array([12, 3, 4])

norm_x = np.linalg.norm(x, ord=2)
norm_y = np.linalg.norm(y, ord=2)

cosf = (x @ y) / (norm_x*norm_y)
math.degrees(np.arccos(cosf))
```

### Задание 3
Определите, будет ли линейное пространство евклидовым, если за скалярное произведение принять:<br>
а) произведение длин векторов;<br>
б) утроенное обычное скалярное произведение векторов?<br>
      
### Решение 3

Для Евклидова пространства должны выполняться 4 аксиомы:<br>
1) $(x,y)=(y,x).$<br>
2) $(\lambda x, y)=\lambda(x, y).$<br>
3) $(x_{1}+x_{2},y)=(x_{1},y)+(x_{2},y).$<br>
4) $(x, x)\geq0$, причём $(x,x)=0\Leftrightarrow x=0.$

a.

По условию:

(x, y) = ||x|| * ||y||

Тогда 3-я аксиома

(||x1 + x2||) * ||y|| = ||x1|| * ||y|| + ||x2|| * ||y||

Берём любые числа для проверки:

x1 = (1, 0)
x2 = (0, 1)
y = (1, 1)

||x1|| = sqrt(1)
||x2|| = sqrt(1)
||x1 + x2|| = sqrt(1+1)
||y|| = sqrt(2)

Подстваляем в уравнение:

sqrt(2)*sqrt(2) = sqrt(1)*sqrt(2) + sqrt(1)*sqrt(2)

sqrt(2) = sqrt(1) + sqrt(1)

sqrt(2) = 2

Это неверное равенство. То есть ответ: нет, это пространство не является Евклидовым

б.

Все аксиомы при умножении на 3 сохраняют равенство.

Ответ: да, это останется Евклидовым пространством.

### Задание 4
Выясните, какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^{3}$:<br>
а) $(1,0,0),(0,0,1);$ <br>
б) $(1/\sqrt{2},-1/\sqrt{2},0),(1/\sqrt{2},1/\sqrt{2},0), (0,0,1);$<br>
в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

### Решение 4

а. Для полноценного базиса $\mathbb{R}^{3}$ не хватает одной координаты.

б. 

```python
v1 = np.array([(1/2)**0.5, -(1/2)**0.5, 0])
v2 = np.array([(1/2)**0.5,  (1/2)**0.5, 0])
v3 = np.array([0, 0, 1])

# Проверяем оргогональность (скларяное произведение должно быть = 0)
print(v1 @ v2)
print(v1 @ v3)
print(v2 @ v3)

# Проверяем нормировку (длина должна быть = 1)
print(v1 @ v1)
print(v2 @ v2)
print(v3 @ v3)

# Все условия выполняются: они ортогональны и нормированы
```
