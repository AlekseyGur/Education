## Задание 1

Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

### Решение 1

Дано:

Нормальное распределение - используем K(a/2) вместо K(a).<br>
sig = 16<br>
p = 0.95<br>
M = 80<br>
n = 256<br>
Найти доверительный интервал<br>

Решение:

Знаем sig, поэтому используем критерий Лапласа. K(a/2) = Z(a/2)<br>
a/2 = 0.05/2 = 0.025<br>
Z(a/2 = 0.025) = -1.96<br>

Интервал = [M - K\*sig/sqrt(n); M - K\*sig/sqrt(n)] = [80 - 1.96\*16/sqrt(256); 80 + 1.96\*16/sqrt(256)] = [78,04; 81,96]

### Ответ 1
[78,04; 81,96] - интервал, в который с вероятностью 95% попадает мат.ожидание

## Задание 2

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:<br>
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1<br>
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

### Решение 2

Дано:

Нормальное распределение - используем K(a/2) вместо K(a).<br>
n = 10<br>
x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]<br>
p = 0.95<br>
a/2 = 0.05/2 = 0.025<br>

Решение:

```python
import numpy as np
x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
np.var(x) # - D(X) несмещённая оценка дисперсии
```

```python
Xavg = sum(x) / len(x)
```

Cигма неизвестна, поэтому для нахождения интервала используем формулу: Xavg +- t(a/2)\*SE, где:<br>
SE = sqrt( D(x)/n )<br>
Xavg - среднее значение по выборке

[Xavg - t(a/2)\*SE; Xavg + t(a/2)\*SE] = [6.59 - 2.262\*sqrt(0.1829/10); 6.59 + 2.262\*sqrt(0.1829/10)] = [6,28; 6,89]

### Ответ2
[6,28; 6,89] - этот интервал с вероятностью 95% покрывает истинное значение величины Х

sig = sqrt(D) = 2

Zрасч = (м1-м0)\*sqrt(n)/sig = (17.5-17)\*10/2 = 2,5<br>

Zрасч > Zтабл, поэтому отвергаем гипотезу Н0
